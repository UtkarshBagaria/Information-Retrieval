```python
def read_files(file_path):
  with open(file_path, 'r',encoding='utf8') as file:
    text = file.read()
  return text
```

Tokens

```python
def tokenize(text):
    tokens = word_tokenize(text)
    return tokens
```

Remove punctuation

```python
import string 
s =  set(string.punctuation) 
s.add('\'')
s.add('n\'t')
# remove punctuation
def remove_punctuation(tokens):
    words = [word for word in tokens if word not in s]
    return words
```

Remove StopWords    

```python
stop_words = set(stopwords.words('english'))
# print(stop_words)
def remove_stopwords(tokens):
    words = [word for word in tokens if word not in stop_words]
    words = [word if word[0]!='\'' else word[1:] for word in words]
    words = [word if word[-1]!='\'' else word[:-1]  for word in words]
    words = [word for word in words if (len(word)>2 and word.isalpha())]
    return words
```

Stemming

```python
# stemming
def stem(token):
    ps = PorterStemmer()
    return ps.stem(token)
# stem(["acknowledge","acknowledgment","acknowledged","acknowledging"])
```

```python
pos_index = {}
bi_wordindex = {}
fileno=0
file_map={}
for file in os.listdir():
  # print(fileno)
  if fileno == 30000:
    break
  if file.endswith('.txt'):
      # Create the filepath of particular file
      file_path =f"{path}/{file}"
      text=read_files(file_path)
      text=text.lower()
      tokens=tokenize(text)
      words=remove_punctuation(tokens)
      words=remove_stopwords(words)
      # print(words)
      # words=lemmatize(words)
      # words=list(set(words))
      # print(words)
      prev_word = ""
      prev_pos = -1
      for pos, term in enumerate(words):
          # First stem the term.
          term = stem(term)
          # Check if the term is in the dataframe.
          # if term in df.columns:
          #     # Increment the presence of the term in the dataframe.
          #     df.loc[df['Document'] == file, term] = 1
          # else:
          #     df[term] = [0]*l
          #     df.loc[df['Document'] == file, term] = 1
          # If term already exists in the positional index dictionary.
          if term in pos_index:
                  
              # Increment total freq by 1.
              pos_index[term][0] = pos_index[term][0] + 1
                  
              # Check if the term has existed in that DocID before.
              if fileno in pos_index[term][1]:
                  pos_index[term][1][fileno].append(pos)
                      
              else:
                  pos_index[term][1][fileno] = [pos]

          # If term does not exist in the positional index dictionary
          # (first encounter).
          else:
                  
              # Initialize the list.
              pos_index[term] = []
              # The total frequency is 1.
              pos_index[term].append(1)
              # The postings list is initially empty.
              pos_index[term].append({})     
              # Add doc ID to postings list.
              pos_index[term][1][fileno] = [pos]

          if prev_word != "":

              bi_word = prev_word + " " + term

              if bi_word in bi_wordindex:

                  bi_wordindex[bi_word][0] = bi_wordindex[bi_word][0] + 1

                  if fileno in bi_wordindex[bi_word][1]:
                      bi_wordindex[bi_word][1][fileno].append(prev_pos)

                  else:
                      bi_wordindex[bi_word][1][fileno] = [prev_pos]

              else:

                  bi_wordindex[bi_word] = []

                  bi_wordindex[bi_word].append(1)

                  bi_wordindex[bi_word].append({}) 

                  bi_wordindex[bi_word][1][fileno] = [prev_pos]

          prev_word = term
          prev_pos = pos
      file_map[fileno] = file
      fileno+=1
# print(file_map)
# print(pos_index)
# write to a file
```

Query-Document Retrieval using Similarity Index with Vector Space Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
doc_vect = {}
fno = 0
for file in list(file_map.values()):
    if file.endswith('.txt'):
        # Create the filepath of particular file
        file_path =f"{path}/{file}"
        text=read_files(file_path)
        text=text.lower()
    
        doc_vect[fno] = text
        fno+=1

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(doc_vect.values())

```

```python
def search_documents(query, n=10):
    query_tfidf = tfidf.transform([query])
    cosine_similarities = cosine_similarity(query_tfidf, tfidf_matrix).flatten()
    related_docs_indices = cosine_similarities.argsort()[::-1][:n]
    return [doc_vect[index] for index in related_docs_indices], related_docs_indices
```

ADVANCED SEARCH: Relevance Feedback and reranking of results

```python
# relevance feedback
def relevance_feedback(tfdf):
    # Prompt the user for feedback
    while(True):
        print("Enter feedback:")
        feedback = input()
        if feedback == "":
            break
        # Process the feedback
        positive_docs = []
        negative_docs = []
        for line in feedback.split("\n"):
            if line.startswith("+"):
                positive_docs.append(int(line[1:]))
            elif line.startswith("-"):
                negative_docs.append(int(line[1:]))
        # print(positive_docs)
        # print(negative_docs)
        for doc in positive_docs:
            if doc in tfdf:
                tfdf[doc]+=1.508456457
        for doc in negative_docs:
            if doc in tfdf:
                if tfdf[doc]>0.898502342:
                    # print("fbijewfhweg")
                    tfdf[doc]-=0.898502342
                else:
                    tfdf[doc]=0
    sorted_dt = {key: value for key, value in sorted(tfdf.items(), key=lambda item: item[1],reverse=True)}
    return sorted_dt
# tf_idf("new york",positional_retrieval("new york"))[0]
relevance_feedback(tf_idf("new york",positional_retrieval("new york"))[0])
    
```
